```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('Dados/housing_ok.csv')
housing.info()
```

```python
housing.plot.scatter(x='longitude', y='latitude', figsize=(10,8))
plt.axis('equal')
```

```python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', figsize=(10,8))
plt.axis('equal')
```

```python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', s=housing.area_construida/5, figsize=(10,8))
plt.axis('equal')
```

```python
housing.plot.scatter(x='longitude', y='latitude', c='preco', cmap='magma', s=housing.area_construida/5, figsize=(10,8), alpha=0.5)
plt.axis('equal')
```

```python
housing.plot.scatter(x='area_construida', y='preco')
```

```python
filtro = (housing.area_construida <= 500) & (housing.preco <= 5000000)
housing[filtro].plot.scatter(x='area_construida', y='preco')
```

```python
housing[filtro].plot.hexbin(x='area_construida', y='preco', cmap='magma', figsize=(10,8))
```
