```python
import pandas as pd
import matplotlib.pyplot as plt

housing = pd.read_csv('Dados/housing_ok.csv')
housing.head()
```

```python
housing.boxplot(column='distancia_aeroporto')
```

```python
housing.boxplot(column='preco', by='tipo')
```
