```python
import pandas as pd

df_countries = pd.read_csv('../Dados/countries.csv')
df_countries.head()
```

```python
df_countries.country.unique()
```

```python
brasil = df_countries[df_countries.country == 'Brazil'].copy()
eua = df_countries[df_countries.country == 'United States'].copy()
argentina = df_countries[df_countries.country == 'Argentina'].copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.plot(brasil.year, brasil.population)
plt.plot(eua.year, eua.population)
plt.plot(argentina.year, argentina.population)
plt.legend(['Brasil', 'EUA', 'Argentina'])
```

```python
sns.catplot(x='year', y='population', kind='bar',data=brasil)
sns.catplot(x='year', y='population', kind='bar',data=eua)
sns.catplot(x='year', y='population', kind='bar',data=argentina)
```

```python
paises = [brasil, eua, argentina]
for i in range(0,len(paises)):
    paises[i]['rel_population'] = paises[i].population/paises[i].iloc[0,2]
```

```python
plt.figure(figsize=(10,5))
for i in range (0, len(paises)):
    temp = paises[i].copy()
    plt.plot(temp.year, temp.rel_population)
plt.legend(['Brasil', 'EUA', 'Argentina'])
```

```python
paises = []
lista = ['Brazil', 'Argentina', 'Colombia', 'Paraguay', 'Uruguay', 'Chile', 'Bolivia', 'Venezuela']
for i in range(0,len(lista)):
    temp = df_countries[df_countries.country == lista[i]].copy()
    paises.append(temp)
```

```python
plt.figure(figsize=(10,5))
for i in range (0, len(paises)):
    temp = paises[i].copy()
    plt.plot(temp.year, temp.population)
plt.legend(lista)
```
