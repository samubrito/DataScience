```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

housing = pd.read_csv('../Dados/housing_ok.csv')
housing.head()
```

```python
filtro = (housing.tipo=='apartamento') & (housing.area_terreno<=1000)
plt.scatter(housing[filtro].area_terreno, housing[filtro].preco)
```

```python
filtro = (housing.tipo=='casa') & (housing.preco<=5000000)
plt.scatter(housing[filtro].latitude, housing[filtro].preco)
```

```python
casas = housing[housing.tipo=='casa']
kitnets = housing[housing.tipo=='kitnet']
apartamentos = housing[housing.tipo=='apartamento']

plt.figure(figsize=(15,10))
plt.axis('equal')
plt.scatter(casas.longitude, casas.latitude)
plt.scatter(kitnets.longitude, kitnets.latitude)
plt.scatter(apartamentos.longitude, apartamentos.latitude)
plt.legend(['casa','kitnet','apartamento'])
```

```python
plt.figure(figsize=(15,10))
plt.axis('equal')
plt.scatter(housing.longitude, housing.latitude, c=housing.distancia_aeroporto)
plt.colorbar()
```

```python
plt.figure(figsize=(15,10))
plt.axis('equal')
plt.scatter(housing.longitude, housing.latitude, c=housing.ano_construcao, cmap='winter')
plt.colorbar()
```
